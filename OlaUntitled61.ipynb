## **Import Libraries**

```python
# Python Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as ex
import warnings
warnings.filterwarnings("ignore")
```

```python
# Data Loadings.
data=pd.read_csv("C:/Users/sachi/Downloads/ola_rides_csb.csv")
print(data)
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
# Handle missing value

df['cancellation_reason']=df['cancellation_reason'].fillna('Not Applicable')
```

```python
df.isnull().sum()
```

```python

df.duplicated().sum()
df=df.drop_duplicates()
```

```python
# Correct Data Types

df['booking_date']=pd.to_datetime(df['booking_date'])
df['booking_hour']=df['booking_hour'].astype(int)
df['fare_amount']=df['fare_amount'].astype(float)
df['ride_distance']=df['ride_distance_km'].astype(float)
```

```python
df['city']=df['city'].str.title().str.strip()
df['vehicle_type']=df['vehicle_type'].str.title().str.strip()
df['payment_method']=df['payment_method'].str.title().str.strip()
```

```python
# Check invalid fare and distance
df=df[(df['fare_amount']>0) & (df['ride_distance_km']>0)]

# Ensure ratings are in valid  range

df = df[(df['driver_rating'] >= 1) & (df['driver_rating'] <= 5)]
df = df[(df['customer_rating'] >= 1) & (df['customer_rating'] <= 5)]
```

## **Outlier Detections**

```python
Q1=df['fare_amount'].quantile(0.25)
Q3=df['fare_amount'].quantile(0.75)
IQR=Q3-Q1
df=df[(df['fare_amount']>=Q1-1.5*IQR)&(df['fare_amount']<=Q3+1.5*IQR)]
```

## **Create Columns**

```python
# Peak vs Non-Peak Hours

df['time_slot'] = np.where(df['booking_hour'].between(6, 9) |
                           df['booking_hour'].between(17, 21),
                           'Peak', 'Non-Peak')
```

## **EDA Analysis**

## **Ride Status Distributions**

```python
plt.figure(figsize=(10,5))
df['city'].value_counts().plot(kind='bar')
plt.title("Ride Status Distribution")
plt.xlabel("Ride Status")
plt.ylabel("Count")
plt.show()
```

## **Hourly Ride Demand Analysis**

```python
hourly_ride_demand=df.groupby('booking_hour').size()
plt.figure(figsize=(10,5))
hourly_ride_demand.plot(kind='line',marker='o')
plt.title("Ride Demand By Hour")
plt.xlabel("Hour Of Days")
plt.ylabel("Number Of Rides")
plt.show()
```

## **City wise Ride Demand**

```python
plt.figure(figsize=(10,5))
df['city'].value_counts().plot(kind='bar')
plt.title("City Wise Ride Demand")
plt.xlabel("City")
plt.ylabel("Nuber of Rides")
plt.show()
```

## **Ride Cancellations Analysis**

```python
cancelled_ride=df[df['ride_status']=='Cancelled']
plt.figure(figsize=(10,5))
cancelled_ride['cancellation_reason'].value_counts().plot(kind='bar')
plt.title("Cancellations Reasons")
plt.xlabel("Reason")
plt.ylabel("Count")
plt.show()
```

## **Vehicle Types vs Ride Count**

```python
plt.figure(figsize=(10,5))
df['vehicle_type'].value_counts().plot(kind='bar')
plt.title("Vehicle Type Usage")
plt.xlabel("Vehicle Type")
plt.ylabel("Number of Rides")
plt.show()
```

## **Fare vs Distance Analysis**


```python
plt.figure(figsize=(10,5))
plt.scatter(df['ride_distance_km'],df['fare_amount'])
plt.title("Fare vs Distance")
plt.xlabel("Distance (KM)")
plt.ylabel("Fare Amount")
plt.show()
```

## **Driver Rating Distributions**

```python
plt.figure(figsize=(10,5))
df['driver_rating'].plot(kind='hist',bins=10)
plt.title("Driver Rating Distributions")
plt.xlabel("Driver Rating")
plt.show()
```

## **Does Low rating cause more cam=ncellations**

## **Revenue of complete rides**

```python
completed = df[df['ride_status'] == 'Completed']

city_revenue = completed.groupby('city')['fare_amount'].sum()

plt.figure()
city_revenue.plot(kind='bar')
plt.title("City-wise Revenue")
plt.xlabel("City")
plt.ylabel("Total Revenue")
plt.show()
```
